```python
class k_mean:
    def __init__(self, k  , date ):
        self.k=k
        self.date =np.array(date)
    
        
    def distance (self,x ,y):
    
       result=np.linalg.norm(np.array(x) - np.array(y))
       return result 
    
    
    def clustring(self,C1,C2): #dividing of pointa 
        cluster1=[]
        cluster2=[]
        
        for i in  self.date :
            if self.distance(C1,i)< self.distance(C2 , i):
                cluster1.append(i)
            else:
                cluster2.append(i)
        return cluster1 , cluster2
    
    def recenter(self,x): #calculation of mean 
         x_mean=0
         y_mean=0
         for points in x :
            x_mean+=points[0]
            y_mean+=points[1]
         x_mean=x_mean/len(x) if len(x)>0 else 0
         y_mean=y_mean/len(x) if len(x)>0 else 0
         return [float(x_mean) , float(y_mean)]
    
        
    def fit(self ,C1,C2): 
      while True:
        CL1,CL2= self.clustring(C1,C2) #cluster 1 and cluster 2 
        old_c1= C1
        old_c2=C2
        C1=self.recenter(CL1)
        C2=self.recenter(CL2)
        if old_c1 == C1 and old_c2 == C2:
             break
      return C1,C2
        
    
    
                
                
            
```

```python
data = [[0,0], [0,1], [-1,2], [2,0], [3,0], [4,-1]]
c1=data[0]
c2=data[-3]
cluster=k_mean(2,data)
new_center1, new_center2=cluster.fit(c1, c2 )
print(f"New Center 1: {new_center1}\nand New Center 2:{new_center2}")
```
