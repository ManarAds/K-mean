```python
import numpy as np

class k_mean:
    def __init__(self, k  , date , tol=0.001, max_iter=300 ):
        self.k=k
        self.date =np.array(date)
        self.max_iter=max_iter
        self.tol=tol
    
        
    def distance (self,x ,y):
       result=np.linalg.norm(np.array(x) - np.array(y))
       return result 
    
    
    
    def fitting (self,date):
        
     self.centroid={}
    
     for i in range (self.k):
         self.centroid[i]=date[i]
         
     for i in range(self.max_iter):
         self.classification={i:[] for i in range(self.k)}
    
         for value in date:
          distance=[self.distance(value , self.centroid[centroids ])for centroids in self.centroid]
          classification=distance.index(min(distance))
          self.classification[classification].append(value)
     prev_centroid=dict(self.centroid)
     for classification in self.classification:
         if len(self.classification[classification]) > 0:
          self.centroid[classification]=np.average(self.classification[classification] , axis=0)
    
     optimize=True
     for centroid in self.centroid:
      orignal_centroid=prev_centroid[centroid]
      currant_centroid=self.centroid[centroid]
      if np.sum((currant_centroid - orignal_centroid) / orignal_centroid * 100 ) > self.tol:
       optimize=False
      if optimize :
         break
     
    
    def predict(self,date):
        distance=[self.distance(date,self.centroid[centroid] )for centroid in self.centroid]
        classification = distance.index(min(distance))
        
        return classification
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib import style
import numpy as np

data = {'x': [25,34,22,27,33,33,31,22,35,34,67,54,57,43,50,57,59,52,65,47,49,48,35,33,44,45,38,43,51,46],
        'y': [79,51,53,78,59,74,73,57,69,75,51,32,40,47,53,36,35,58,59,50,25,20,14,12,20,5,29,27,8,7]}

new_date=np.array(list(zip(data['x'], data['y'])))

style.use('ggplot')
X = new_date

plt.scatter(X[:,0], X[:,1], s=150)
plt.xlabel("the x co-ordnate")
plt.ylabel('the y co-ordinate')
plt.show()


```

```python

clf=k_mean(k=3 , date=new_date)
clf.fitting(new_date)

num_clusters = len(clf.classification)
colors = plt.cm.rainbow(np.linspace(0, 1, num_clusters)) 
color_map = {i: colors[i] for i in range(num_clusters)} 


plt.figure(figsize=(8, 6))
for c in clf.centroid:
  plt.scatter(clf.centroid[c][0] , clf.centroid[c][1] , marker='o' , c='k' , s=150 , linewidths=3)
  
  
for classification in clf.classification:
    color = color_map[classification]
    for value in clf.classification[classification]:
        plt.scatter(value[0], value[1], marker="x", color=color, s=150, linewidths=3)
        
plt.show()
```
